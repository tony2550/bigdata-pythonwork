```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
def content(url):
    res=urlopen(url)
    soup=bs(res, 'html.parser')
    cont=soup.select('div.story_area > p')
    con='19세 이상만'
    if len(cont)>0:
        con=cont[0].text
        #print(con)
    return con

def sub_craw(url):
    res=urlopen(url)
    soup=bs(res,'html.parser')
    temp=soup.select('#content > div.article > div.mv_info_area > div.mv_info > div.viewing_graph')
    if len(temp)>0:
        temp=temp[0].text.replace("\n","").replace('성별·나이별 관람추이남자여자','').replace('대 이상','').split('대')
        trend= '10대' + temp[0].replace('10','') + ' 20대' + temp[1].replace('20','') \
            + ' 30대' + temp[2].replace('30','') + ' 40대' + temp[3].replace('40','') + ' 50대' + temp[4].replace('50','')
        return trend
# print(sub_craw('https://movie.naver.com/movie/bi/mi/basic.naver?code=35360'))
```

```python
import requests #이름_데이터수집_과제1 현재 상영영화
from bs4 import BeautifulSoup as bs
url='https://movie.naver.com/movie/running/current.naver'
res=requests.get(url)
soup=bs(res.text,'html.parser')
#print(soup)
movies=soup.select('#content > div.article > div:nth-child(1) > div.lst_wrap > ul > li')
#print(movies)
i=0
movie_list=[]
url1='https://movie.naver.com'
for movie in movies:
    #title
    title=movie.select('dl.lst_dsc > dt.tit > a')[0].string
#     print(title)
    #image
    image=movie.select('div > a > img')
    src=image[0]['src']
    #print(img)
#     print(src)
    image_save(src,i)
    #rating
    rating_select=movie.select('dl.lst_dsc > dt.tit > span')
    if len(rating_select)>0:
        rating=rating_select[0].string
#     print(rating)
    #star
    star=movie.select('dl > dd.star > dl.info_star > dd > div > a > span.num')[0].string
#     print(star)
    #genre ,running_time , opening_day
    movie_info=movie.select('dl.info_txt1 > dd:nth-child(2)')[0].text.replace('\t','').replace('\n','').replace('\r','').split('|')
#     print(movie_info)
    if len(movie_info)>2:
        genre=movie_info[0]
        running_time=movie_info[1]
        opening_day=movie_info[2]
    else:
        genre=''
        running_time=movie_info[0]
        opening_day=movie_info[1]
#     print(genre)
#     print(running_time)
#     print(opening_day)
    #director
    director=movie.select('dl.info_txt1 > dd:nth-child(4) > span.link_txt > a')[0].string
#     print(director)
    #actor
    actor_select=movie.select('dl.info_txt1 > dd:nth-child(6) > span.link_txt')
    if len(actor_select)>0:
        actor=actor_select[0].text.replace('\t','').replace('\n','').replace('\r','')
#         print(actor)
    tit_a=movie.select('dl > dt.tit')
    tit_href=tit_a[0].a['href']
#     print(tit_href)
#     print(url1+tit_href)
    summary=content(url1+tit_href)
    viewing_trend=sub_craw(url1+tit_href)
    print(title)
    movie_list.append([title,rating,star,genre,running_time,opening_day,director,actor,summary,viewing_trend])
    i+=1
print(movie_list)
```
